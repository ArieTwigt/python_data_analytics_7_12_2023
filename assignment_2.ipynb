# Assignment 5/6

Group the DataFrame based on the purpose column. The value should be the average credit amount per purpose.

```python
select_columns = ['purpose', 'credit_amount']

data_grouped = (
        data[select_columns]
        .groupby('purpose')
        .mean()
        .rename(columns = {"credit_amount": "mean_credit_amount"})
        .sort_values(by="mean_credit_amount", ascending=False)
        .reset_index() # assignment 6
        #.query("mean_credit_amount > 5000")
        #.assign(mean_credit_amount_dollars = lambda x: x['mean_credit_amount'] * 0.9)
)

data_grouped
```

## Assignment 7

Export the table to the sqlite database (data/data.db) to a table named large_loans_purposes

```python
# specify the path for the sqlite database
db_path = "data/data.db"

# establish the connection to the database
con = sqlite3.connect(db_path)
```

```python
# use pandas to export to the database
data_grouped.to_sql("large_loans_purposes", con=con, if_exists='replace')
```
