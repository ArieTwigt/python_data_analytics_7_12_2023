```python
# import the require modules
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score
```

```python
# import the dataset
data = pd.read_csv("data/german_data_clean.csv")
data.head()
```

```python
# show the division of the response
data['response'].value_counts()
```

```python
# specify what you would like to predict
X_columns = ['age_years', 'duration_months', 'credit_amount']
y_column = 'response'
```

```python
# based on the values, get the right columns
X = data[X_columns]
y = data[y_column]
```

```python
# split the data in a train- and test set
X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=0.3)
```

```python
# choose the algorithm to train
model = LogisticRegression()
```

```python
# train the model
model.fit(X_train, y_train)
```

```python
# generate predictions probabilities
model.predict_proba(X_test)
```

```python
# generate the predictions
predictions = model.predict(X_test)
```

```python
# compare the predictions with the real values in the test-set
df_comparison = pd.DataFrame()

# 
df_comparison = X_test
df_comparison['y_actual'] = y_test
df_comparison['y_pred'] = predictions
df_comparison
```

```python
# score the model : accuracy
accuracy_score(df_comparison['y_actual'], df_comparison['y_pred'])
```

```python
# # score the model : precision
precision_score(df_comparison['y_actual'], df_comparison['y_pred'])
```
